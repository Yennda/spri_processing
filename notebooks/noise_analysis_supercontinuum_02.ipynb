```python
folder = r'D:\SPRUP_data_Jenda\21_04_14_DG750A1/'.replace('\\', '/')

file = 'raw_01_'
i = 3

c = Core(folder, file+str(i))
c.k = 10
c.type = 'diff'
c._data_raw = c._data_raw[:60, :200, :208]

# 100ms x 1
```

```python
c_std = np.zeros(c.shape_img)

for i in range(c.shape_img[0]):
    for j in range(c.shape_img[1]):    
        c_std[i, j] = np.std(c._data_raw[i, j, :])
```

```python
d_std = []
for f in range(20, len(c)):
    d_std.append(np.average(np.abs(c.frame(f))))
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
# d =  c._data_raw[:, :, 50]
d =  c_std[:, :]

axes.imshow(
    d,
    cmap='gray',
    zorder=0,
#     vmin=np.min(d)*0.2,
#     vmax=np.max(d)*0.2
)
```

```python
i_avg = np.average(c._data_raw)
print('px std: {}'.format(np.average(c_std)))
print('diff frame std: {}'.format(np.average(d_std)))
print('avg int: {}'.format(i_avg))
print('-'*20)
print('px std: {}'.format(np.average(c_std)/i_avg))
print('diff frame std: {}'.format(np.average(d_std)/i_avg))
```
