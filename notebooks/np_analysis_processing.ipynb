```python
main_folder = r'C:\SPRUP_data_Jenda\2019_03_13_Jenda_microscopy'.replace('\\', '/')

folders = listdir(main_folder)

for folder in folders:
    m_folders = [f for f in listdir(main_folder + '/' + folder) if f.find('.') == -1]
    
    if 'saved_data' in m_folders:
        sd_folder = listdir(main_folder + '/' + folder + '/saved_data') 
        print('\t' + '='*20)
        print('\t►► ' + folder)
        print('\t' + '-'*20)
        print('area, intensity, intensity_px, intensity_bg_px, snr, success/count\n')
        
        for file in sd_folder:
            
            if re.search('^np_analysis.*', file):
                w, h, frame_time, avg, num_of_frames, ets = tl.read_file_info(main_folder + '/' + folder + '/' + file[12:-4])
                
                print('\n\t\t♥ '+ file + '; {} x {} ms'.format(avg, ets) + '\n')
                print(ets, avg)
                
                with open(main_folder + '/' + folder + '/saved_data/' + file, mode='r') as csv_file:
                    csv_reader = csv.reader(csv_file, delimiter=',')
                    line_count = 0
                    
                    for i, row in enumerate(csv_reader):
                        if i == 1:
                            values = [float(r) for r in row]
                            values[1:4] = [v/ets for v in values[1:4]]
                        if i == 2:
                            stds = [float(r) for r in row]
                            stds[1:4] = [s/ets for s in stds[1:4]]
                            
                    print(values)
                    print(stds)
```
