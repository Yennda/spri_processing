```python
folder = r'C:\SPRUP_data_Jenda\2022_07_19_UV\22_08_29/'.replace('\\', '/')

file = 'raw_13_1'


core = Core(folder, file)
core.type = 'int'
print(len(core))
```

```python
mask = np.zeros(core.shape_img)

measurement = []

slices_x = []
slices_y = []
slices_ref = []

for j in range(4):
    k = 3 - j
    for i in range(3):
        dx = 40 + i * 217
        dy = 70 + k * 180
        sx = slice(dx, dx + 95)
        sy = slice(dy, dy + 95)
        slices_x.append(sx)
        slices_y.append(sy)
        
        measurement.append((i, j))
        mask[sx, sy] = 1
        
    dy = 70 + k * 180
    sr = slice(dy, dy + 95)

    slices_ref.append(sr)

masks_r = []
for k in range(4):
    mask_r = np.zeros(core.shape_img)
    for i in range(4):
        if i != k:
            mask_r[:, slices_ref[i]] = 1
    for j in range(3):    
        masks_r.append(mask_r == 1)
```

```python
doses = [56.8, 30.3, 14.9, 10.0, 5.0, 2.1]
doses_d =[]
for d in doses:
    doses_d.append(d)
    doses_d.append(d)
    
doses = np.array(doses)
doses_d = np.array(doses_d)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.set_title('Response of oligonucleotide to UV illumination')
# axes.set_suptitle('as')

axes.set_xlabel('time [min]')
axes.set_ylabel('Sensor response [a.u.]')

colors = pl.cm.jet(np.linspace(0,1,13))

responses = []

for i in range(1, 13):

    file = 'raw_{:02d}_1'.format(i)
    core = Core(folder, file)
    core.type = 'raw'
    print(len(core))
    
    time = np.array([i*2.8/60 for i in range(1, len(core))])


    spr = []
    for f in range(1, len(core)):
        spr.append(np.average(core.frame(f)[slices_x[i - 1], slices_y[i - 1]]))
    spr = np.array(spr)
    spr = spr / np.average(spr[:85]) * (-1)


    spr_ref = []
    for f in range(1, len(core)):
        spr_ref.append(np.average(core.frame(f)[masks_r[i - 1]]))
    spr_ref = np.array(spr_ref)
    spr_ref = spr_ref / np.average(spr_ref[:85]) * (-1)




    # axes.plot(spr)
    # axes.plot(spr_ref)
    signal = spr - spr_ref
    axes.plot(time, signal, label = '{} mW/cm$^2$'.format(doses_d[i - 1]), color=colors[i - 1])
    
    responses.append(np.average(signal[350:]))

lgd = fig.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))

# fig.savefig('images/'+ file +'.png', dpi=300, bbox_inches='tight')    
# fig.savefig(folder +'response_curves.png', dpi=300, bbox_inches='tight')    

```

```python



responses_avg =[]
for i in range(len(doses)):
    responses_avg.append(np.average(responses[i*2:i*2+2]))

responses_avg = np.array(responses_avg)

p = np.polyfit(np.log(doses), responses_avg, 1)

a = p[1]
b = p[0]

doses_fitted = np.linspace(np.min(doses), np.max(doses), 100)

# responses_fitted = b*np.log(doses_fitted) + np.log(a)
# responses_fitted = b*np.log(doses_fitted) + a
responses_fitted = np.log(np.exp(b)*doses_fitted**a)

    
    
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.set_title('Response of oligonucleotide to UV illumination')
axes.set_xlabel('dose [mWcm$^{-2}$ ]')
axes.set_ylabel('sensor response [a. u.]')
# axes.set_xlim(60, 0)
# axes.set_ylim(0, 0.2)

axes.scatter(doses_d,responses, label = 'experimental data')
axes.scatter(doses, responses_avg, color = 'red', label = 'average of ex. data')
axes.plot(doses_fitted, responses_fitted, color = 'black', ls = '--', label = 'fitted curve')


textstr = '$y = ln(bx^a)$\n$b = {:.5f}$, \n$a = {:.5}$'.format(np.exp(b), a)
props = dict(boxstyle='square', facecolor='white', alpha=0.5)
axes.text(0.6, 0.3, textstr, transform=axes.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)


# axes.scatter(np.log(doses), responses_avg, color = 'red')
# # axes.scatter(np.log(doses), responses_avg, color = 'red')

# # axes.plot(np.log(doses_fitted), b*doses_fitted + a, color = 'black')

lgd = fig.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))


# fig.savefig(folder +'response_fit.png', dpi=300, bbox_inches='tight')    


```
